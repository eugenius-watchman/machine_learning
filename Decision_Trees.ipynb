```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.metrics import ConfusionMatrixDisplay
#from sklearn.metrics import plot_confusion_matrix ...deprecated

%matplotlib inline
```

```python
DATAPATH = "data/breastCancer.csv"

data = pd.read_csv(DATAPATH)

data.head()
```

```python
# check if dataset is balanced
target_column = data["Classification"]
ax = sns.countplot(x=target_column, data=data)

ax.set_title("Class Distribution")  # Add title
ax.set_xlabel("Cancer Diagnosis")   # Customize x-label
ax.set_ylabel("Count")     
```

```python
# distribution of each feature for both classes
def violin_plots(x, y, data):
    for i, col in enumerate(y):
        plt.figure(i)
        sns.set(rc={"figure.figsize":(11.7, 8.27)})
        ax = sns.violinplot(x=x, y=col, data=data)
        plt.title(f"{col} distribution by {x}")
        plt.show()

y = data.columns[:-1]
x = data.columns[-1]

violin_plots(x, y, data)
```

```python
# check for null values ... make sure nothing is missing
for col in data.columns:
    print(f"{col}: {data[col].isnull().sum()}")
```

```python
## Preprocessing 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data["Classification"] = le.fit_transform(data["Classification"])

data.head()
```

```python
# split data into training and test set
from sklearn.model_selection import train_test_split

y = data["Classification"].values.reshape(-1, 1)
X = data.drop(["Classification"], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
# build Baseline Model

from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()

clf.fit(X_train, y_train)

# plot_confusion_matrix(clf, X_test, y_test, cmap=plt.cm.Blues) #deprecated
# plot confusion matrix
ConfusionMatrixDisplay.from_estimator(
    clf, X_test, y_test,
    cmap=plt.cm.Blues,
    display_labels=["Healthy", "Cancer"]
)


plt.grid(False)
plt.show()
```

```python
# visualize decision trees with the function plot tree

from sklearn.tree import plot_tree

plot_tree(clf, max_depth=5, filled=True)
```

```python
# improve on baseline model using bagging
from sklearn.ensemble import BaggingClassifier

bagging_clf = BaggingClassifier()

bagging_clf.fit(X_train, y_train.ravel())

# plot_confusion_matrix(bagging_clf, X_test, y_test, cmap=plt.cm.Blues) # deprecated
# plot confusion matrix
ConfusionMatrixDisplay.from_estimator(
    bagging_clf, X_test, y_test,
    cmap=plt.cm.Blues,
    display_labels=["Healthy", "Cancer"] #optional
)

plt.grid(False)
plt.show()
```

```python
# implementing Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

randomF_clf = RandomForestClassifier(100)

randomF_clf.fit(X_train, y_train.ravel())

ConfusionMatrixDisplay.from_estimator(
    randomF_clf, X_test, y_test,
    cmap=plt.cm.Blues
)
plt.grid(False)
plt.show()
```

```python
## implement Boosting

from sklearn.ensemble import GradientBoostingClassifier

GradBoosting_clf = GradientBoostingClassifier()

GradBoosting_clf.fit(X_train, y_train.ravel())

# plot confusion matrix 
ConfusionMatrixDisplay.from_estimator(
    GradBoosting_clf, X_test, y_test,
    cmap=plt.cm.Blues,
    display_labels=["Healthy", "Cancer"] #optional
)
plt.grid(False)
plt.show()
```
