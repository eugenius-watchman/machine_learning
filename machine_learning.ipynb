```python
from sklearn.datasets import load_iris

iris = load_iris()

data = iris.data

target = iris.target
```

```python
data
```

```python
target
```

```python
# understand structure of dataset
print("Feature Names", iris.feature_names)
```

```python
# information about flowers
print("Target Name", iris.target_names)
```

```python
# data shape
print("Data shape", data.shape)
```

```python
 # convert dat to pandas dataframe
import pandas as pd
df = pd.DataFrame(data, columns = iris.feature_names)

df
```

```python
# set target
df["target"] = target

df
```

```python
# Clean data ...check for missing values
df.isnull().sum()
```

```python
# normalisation
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_data = scaler.fit_transform(data)

normalised_data
```

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
scaled_data = standard_scaler.fit_transform(data)

scaled_data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.2)
```

```python
# for regression and classification
# import scoring metrics
from sklearn.metrics import accuracy_score, recall_score, f1_score, confusion_matrix
from sklearn.metrics import precision_score

from sklearn.neighbors import KNeighborsClassifier

# initialise model class
knn = KNeighborsClassifier(n_neighbors = 2)

# fitting the model
knn.fit(X_train, y_train)

# make predictions based on X-test and compare performances with y_test
predictions = knn.predict(X_test)

predictions
```

```python
y_test
```

```python
accuracy_score(y_test, predictions)
```

```python
 # precision
precision_score(y_test, predictions, average = "weighted")
```

```python
## recall_score
recall_score(y_test, predictions, average = "weighted")
```

```python
#f1_score
f1_score(y_test, predictions, average = "weighted")
```

```python
confusion_matrix(y_test, predictions)
```

```python
## Regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


iris = load_iris()
data = iris.data

y = data[:, 0]
y
```

```python
# Work on Regression Model
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size = 0.2)
```

```python
y_test
```

```python
y_train
```

```python
X_train
```

```python
# Create a regression model
reg_model = LinearRegression()

reg_model.fit(X_train, y_train)
```

```python
# predictions
predictions = reg_model.predict(X_test)
```

```python
# mse 
mean_squared_error(y_test, predictions)
```

```python
# r2_score
r2_score(y_test, predictions)
```
