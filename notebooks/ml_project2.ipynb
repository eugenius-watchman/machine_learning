```python
import pandas as pd
```

```python
# reading data
data = pd.read_csv("data/Credit_card.csv")
data
```

```python
# merge dataframes
label = pd.read_csv("data/Credit_card_label.csv")
label

```

```python
# merge on Ind_ID ... using a Left Join
data = data.merge(label, on = "Ind_ID", how = "left")
data
```

```python
# calculate the avg income 
avg_income = data["Annual_income"].mean()
avg_income
```

```python
gender_income = data.groupby("GENDER")["Annual_income"].mean()
gender_income
```

```python
#gender_income.sort_values(ascending = False)[0]
gender_income.sort_values(ascending = False).head(1)
```

```python
# data visualisation
# plot avg income by gender

sns.barplot(x = gender_income.index, y = gender_income.values)
plt.title("Average Income by Gender")
plt.xlabel("Gender")
plt.ylabel("Money")
plt.show()
```

```python
# calculate value count for label 
label_counts = data["label"].value_counts()
label_counts
```

```python
# make info more meaningful by creating a piechart 
plt.pie(label_counts, labels = ["Credit Approved", "Credit Rejected"], autopct = "%1.1f%%")
plt.title("credit Approval Status")
plt.show()
```
