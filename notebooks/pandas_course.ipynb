```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# dataset
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.2, 100)

# plot points
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
categories = ["A", "B", "C","D","E"]
values = [15, 7, 10, 12, 8]

plt.bar(categories, values)
plt.title("Bat Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
plt.hist(y, bins = 20)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
labels = ["Category A", "Category B", "Category C"]
sizes = [30, 20 , 40]
explode = [0.1, 0, 0]

plt.pie(sizes, explode = explode, labels = labels)
plt.title("Pie Chart")
plt.show()
```

```python
# area plot
plt.fill_between(x, y, alpha = 0.5)
plt.title("Area Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
ohlc = np.vstack((x,y,y,y)).T
plt.plot(x, y, color = "b")
plt.title("Candlestick Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.stem(x, y, "r", markerfmt = "ro")
plt.show()
```

```python
plt.violinplot(y, showmedians = True)
plt.title("Violin Plot")
plt.ylabel("Values")
plt.show
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.plot_trisurf(x, y, x**2 + y**2, cmap = "viridis")
ax.set_title("3D Surface Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()

```

```python
plt.hexbin(x, y, gridsize=20, cmap="inferno")
plt.title("Hexbin Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar(label = "Count")
plt.show()
```

```python
# polar plot
theta = np.linspace(0, 2* np.pi, 100)
r = 2 + np.sin(5 * theta)

plt.polar(theta, r)
plt.title("Polar Plot")
plt.show()
```

```python
x, y = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
u = -y
v = x

plt.streamplot(x, y, u, v)
plt.title("Stream Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
## errorbar plot
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.2, 100)
errors = abs(np.random.normal(0, 0.2, 100))

plt.errorbar(x,y, yerr = errors, fmt = "o", capsize = 3, ecolor = "r")
plt.title("ErrorBar Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
