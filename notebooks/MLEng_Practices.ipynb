```python
class Shirt:
    def __init__(self, shirt_color, shirt_size, shirt_style, shirt_price):
        self.color = shirt_color
        self.size = shirt_size
        self.style = shirt_style
        self.price = shirt_price

    def change_price(self, new_price):
        self.price = new_price

    def discount(self, discount):
        return self.price * (1 - discount)
```

```python
Shirt("red", "S", "short sleve", 15)
```

```python
new_shirt = Shirt("red", "S", "short sleve", 15)
```

```python
print(new_shirt.color)
print(new_shirt.size)
print(new_shirt.style)
print(new_shirt.price)
```

```python
new_shirt.change_price(10)
print(new_shirt.price)
```

```python
print(new_shirt.discount(0.2))
```

```python
tshirt_collection = [] 
shirt_one = Shirt("orange", "M", "short sleeve", "10")
shirt_two = Shirt("green", "XL", "long sleeve", "15")
shirt_three = Shirt("yellow", "L", "short sleeve", "20")
shirt_four = Shirt("purple", "X", " sleeveless", "10")

tshirt_collection.append(shirt_one)
tshirt_collection.append(shirt_two)
tshirt_collection.append(shirt_three)
tshirt_collection.append(shirt_four)

for i in range(len(tshirt_collection)):
    print(tshirt_collection[i].color)
```
