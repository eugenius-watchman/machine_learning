```python
## Resampling and Regularization

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
DATAPATH = "data/advertising.csv"

data = pd.read_csv(DATAPATH)
data.head()
```

```python
# plotting target againt each feature
def scatter_plot(feature):
    plt.figure(figsize=(10,5))
    plt.scatter(data[feature], data["sales"], c= "black")
    plt.xlabel(f"Money spent on {feature} adds ($)")
    plt.ylabel("Sales (k$)")
    plt.show()
```

```python
## plt each feature
features_to_plot = ["TV", "radio", "newspaper"]

for feature in features_to_plot:
    scatter_plot(feature)
    print(f"Plotted:{feature}")
```

```python
# define Baseline model 
## how Regularization will improve it 

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
X = data.drop(["sales"], axis = 1)
y = data["sales"].values.reshape(-1,1)

## fit the model
lin_reg = LinearRegression()
 
# cv=5 => 5 fold cross validation ...give us 5 different mse
MSEs = cross_val_score(lin_reg, X, y, scoring="neg_mean_squared_error", cv = 5)

mean_MSE = np.mean(MSEs)

print(f"Negative MSEs from 5-fold CV: {MSEs}")
print(f"Mean Negative MSE: {mean_MSE}")
print(-mean_MSE)
```

# Regularisation to improve baseline
## Ridge Regression


```python
# Regularisation to improve baseline
## Ridge Regression
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()

# define list of possible values for Tuning parameter
parameters = {"alpha": [1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10 , 20]}

# run 5fold cross validation
ridge_regressor = GridSearchCV(ridge, parameters, scoring="neg_mean_squared_error", cv=5)

ridge_regressor.fit(X, y);

# best value for paramerters and mse
print(ridge_regressor.best_params_)
print(-ridge_regressor.best_score_) #MSE

print(f"\nBest alpha found: {ridge_regressor.best_params_['alpha']}")
print(f"Best score (Negative MSE): {ridge_regressor.best_score_:.4f}")
print(F"Actual MSE: {-ridge_regressor.best_score_:.4f}")
```

```python
#LASSO

from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(tol=0.05)

parameters = {"alpha": [1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10, 20]}

lasso_regressor = GridSearchCV(lasso, parameters, scoring="neg_mean_squared_error", cv=5)

lasso_regressor.fit(X, y);

# best value for paramerters and mse
print(lasso_regressor.best_params_)
print(-lasso_regressor.best_score_) #MSE

print(f"\nBest alpha found: {lasso_regressor.best_params_['alpha']}")
print(f"Best score (Negative MSE): {lasso_regressor.best_score_:.4f}")
print(F"Actual MSE: {-lasso_regressor.best_score_:.4f}")
```
