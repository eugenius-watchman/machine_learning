```python
##Unsupervised Learning

import numpy as np 
import matplotlib.pyplot as plt
from sklearn.utils import shuffle 
```

```python
## Clustering: K-means
from sklearn.datasets import load_sample_image
from sklearn.cluster import KMeans
```

```python
flower = load_sample_image("flower.jpg")

flower = np.array(flower, dtype=np.float64) / 255

plt.imshow(flower);
```

```python
# change image to a 2D matrix 
w, h, d = original_shape = tuple(flower.shape)
assert d == 3
image_array = np.reshape(flower, (w*h, d))
```

```python
# reduce number of colours to 64 by running the KMeans alg 
image_sample = shuffle(image_array, random_state=42)[:1000]

# Fit Kmeans
n_colours = 64  #play around with the number 
kmeans = KMeans(n_clusters=n_colours, random_state=42).fit(image_sample)

# Get colour indices for full image 
labels = kmeans.predict(image_array)

```

```python
# function to rebuild the image
def reconstruct_image(cluster_centers, labels, w, h):
    d = cluster_centers.shape[1]
    image = np.zeros((w, h, d))
    label_index = 0 
    for i in range(w):
        for j in range(h):
            image[i][j] = cluster_centers[labels[label_index]]
            label_index += 1
    return image
```

```python
plt.figure(1)
plt.clf()
plt.axis("off")
plt.title("Original image with 96 615 colours")
plt.imshow(flower);

plt.figure(2)
plt.clf()
plt.axis("off")
plt.title(f"Reconstructed image with {n_colours} colours")
plt.imshow(reconstruct_image(kmeans.cluster_centers_, labels, w, h));
```

```python
# Principal Component Analysis (PCA)
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names 
```

```python
# initialise PCA
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
```

```python
# print out amount of varience
print(f"Explained varience ratio from PCA: {pca.explained_variance_ratio_}")
```

```python
# plot dataset in 2D
colors = ["#003f5c", "#bc5090", "#ffa600"]
lw = 2 #length-width
```

```python
plt.figure()

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y==i, 0], X_r[y==i, 1], color=color, alpha=0.8, lw=lw, label=target_names)

plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("PCA of Iris Dataset")
    
```
